```python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
a=SentimentIntensityAnalyzer()
df=pd.read_csv('tripadvisor_hotel_reviews.csv')
df['Sentiment']=df['Review'].apply(lambda x: 'positive' if a.polarity_scores(str(x))['compound']>=0 else 'negative')
plt.figure(figsize=(5,5))
b=sns.countplot(x='Sentiment', data=df, palette='Accent')
b.bar_label(b.containers[0])
plt.xlabel('no. of reviews')
plt.show() # As we analyze the no of positive reviews 18874 and no. of negative reviews is 1617. The rate of positive reviews is morw
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
a=SentimentIntensityAnalyzer()
df=pd.read_csv('tripadvisor_hotel_reviews.csv')
df['Sentiment']=df['Review'].apply(lambda x: 'positive' if a.polarity_scores(str(x))['compound']>=0 else 'negative')
plt.figure(figsize=(5,5))
sns.barplot(x='Sentiment', y='Rating', data=df, palette='Accent')
plt.xlabel('no. of reviews')
plt.show()  #The positive reviews rating are high as compare to negative reviews
```
